```python
#Importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##Load the dataset
data= pd.read_csv("C://Users/shiva/Downloads/dec18/batch 68/4. Logistic Regresssion - Assignment - 3/4. Logistic Regresssion/bank-full.csv",sep = ';')
data.head(2)
```

```python
##Descriptive stats 
data.describe()
```

```python
##EDA : Data Pre-processing
## Remove Null Values
data.isnull().sum()
```

```python
##This shows there are no missing values !! 
##Removing outliers. 
##Plot the box plot for each numeric column to find the outlier.
sns.boxplot(data=data,x='duration')
```

```python
##To eliminate the outliers found using box blot, you can use the quartile method. 
#Removing the outliers
q3 = data['duration'].quantile(.75)
q1 = data['duration'].quantile(.25)
iqr = q3-q1
iqr
upperrange = q3+1.5*iqr
bottomrange = q1-1.5*iqr
df1 = data[(data['duration']>bottomrange) & (data['duration']<upperrange)] 
sns.boxplot(data=df1,x='duration',color = 'purple')
```

```python
#shows the no.of customers vs Client's Term Deposit Subscription Status
plt.subplots(figsize=(12,4))
sns.countplot(x ='job', hue = 'y', data = df1, palette = 'colorblind')
plt.ylabel('No. Of Customers')
plt.show
```

```python
##Label Encoding 
from sklearn.preprocessing import LabelEncoder
for column in df1.columns:
        if df1[column].dtype == np.number:
            continue
        df1[column] = LabelEncoder().fit_transform(df1[column])
df1.head()
```

```python
#correlation visualization
df1.corr()
plt.subplots(figsize=(12,10))
sns.heatmap(df1.corr(), annot = True)
plt.show
```

```python
#split data into 70% training and 30% testing.
from sklearn.model_selection import train_test_split
X = df1.iloc[:, 1:df1.shape[1]].values
Y = df1.iloc[:,0].values
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3, random_state = 1)
X_train
```

```python
#Logistic Regression model building
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(X_train,Y_train)
```

```python
#Training Model Score
logmodel.score(X_train,Y_train)
```
